```python
import pandas as pd
import numpy as np
import csv
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

**Cargando Datos...**

```python
df=pd.read_csv('regLin.csv')
df.head()
```

**Data Partition con 75% de datos para entrenar**

```python
X_train,X_test,y_train,y_test=train_test_split(df[['X']],df[['y']],test_size=.25)
```

**Ajustando una regresi√≥n Train**

```python
plt.scatter(X_train,y_train)
```

```python
linReg=LinearRegression()
linReg.fit(X_train,y_train)
```

```python
w0=linReg.intercept_[0]
w1=linReg.coef_[0][0]
print(w0,w1)
```

**Test model: Error cuadratico medio**

```python
ECM=np.mean((linReg.predict(X_test)-y_test)**2)
print(ECM)
```

**GRAFICANDO DATOS Y RESULTADO**

```python
x = np.linspace(float(X_test.min()),float(X_test.max()),500) # 100 numeros espaciados
plt.plot(x,x*w1+w0,color='red')
plt.scatter(X_test,y_test,alpha=0.6)
```

### Haga un gr√°fico de como cambia el error al cambiar las w¬¥s. Haga uno para w0 y uno para w1. Ponga un punto rojo para los valores encontrados por la regresi√≥n lineal

```python
W0=np.linspace(w0-1.5,w0+1.5,100)
err=[np.mean((i+w1*X_train-np.array(y_train))**2) for i in W0]
plt.scatter(W0,err)
plt.plot(w0,np.mean((w0+w1*X_train-np.array(y_train))**2),'p-',color='red',ms=9)
```

**Para w1...**

```python
W1=np.linspace(w1-1,w1+1,100)
err=[np.mean((w0+i*X_train-np.array(y_train))**2) for i in W1]
plt.scatter(W1,err)
plt.plot(w1,np.mean((w0+w1*X_train-np.array(y_train))**2),'p-',color='red',ms=9)
```
